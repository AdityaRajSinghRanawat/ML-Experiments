```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('../data/data.csv')
```

```python
# Separate features (X) and target (y)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
def naive_bayes_predict(data, target, test):
    classes = target.unique()
    probs = {}

    for c in classes:
        # 1. Calculate Prior Probability: P(Class)
        subset = data[target == c]
        prior_prob = len(subset) / len(data)
        
        # Start with the prior
        likelihood = prior_prob
        
        # 2. Calculate Likelihood: P(Feature|Class) for each feature in test
        for col, val in test.items():
            # Count occurrences of feature value within the class subset
            count_feature_in_class = len(subset[subset[col] == val])
            # Conditional probability: P(x|c)
            likelihood *= (count_feature_in_class / len(subset))
            
        probs[c] = likelihood

    # Return the class with the maximum probability
    return max(probs, key=probs.get)
```

```python
# Test Sample
test_sample = {
    "Outlook": "Rainy",
    "Temp": "Mild",
    "Humidity": "Normal",
    "Windy": "t"
}
```

```python
# Run Prediction
prediction = naive_bayes_predict(X, y, test_sample)
print(f"Prediction for Play: {prediction}")
```
