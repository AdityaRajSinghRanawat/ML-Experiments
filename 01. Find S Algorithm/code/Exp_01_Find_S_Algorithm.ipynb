Python slicing syntax used primarily with NumPy arrays (and lists of lists)

* :   => before the comma means 'select all rows'.

* :-1 => after the comma means 'select all columns except the last one'. The -1 index refers to the last element, so :-1 means up to, but not including, the last element.

```python
concepts = np.array(data)[:,:-1]
```

```python
concepts
```

[:, -1] means all rows but only last column

```python
target = np.array(data)[:,-1]
```

```python
target
```

```python
def train(con, tar):
    for i, val in enumerate(tar): # i = 0 .... n, val = yes/no
        if val == 'yes':
            specific_h = con[i].copy() # set "h" value to current con and make a copy of con to do not modify it
            break

    for i, val in enumerate(con): # i = 0 ....n, val = array
        if tar[i] == 'yes':
            for x in range(len(specific_h)): # x = 0 to end of row
                if val[x] != specific_h[x]:
                    specific_h[x] = '?'
                else:
                    pass
    return specific_h
```

```python
# function calling
print(train(concepts, target))
```
